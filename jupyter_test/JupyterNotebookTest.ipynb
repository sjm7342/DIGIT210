```python
with open("avatarSpeeches.py") as file:
    br = file.read()

print(type(br))
print(len(br))

howLong = len(br)
print(f"howLong = {howLong}")

novelSlice = br[:500]
print(f"novelSlice = {novelSlice}")

splitEmUp = br.split()
print(f"splitEmUp = {splitEmUp[-100:]}")

for token in splitEmUp:
    if token.endswith('ing'):
        print(token)
```
